```python
'''
Q23. Generate synthetic data using make_moons and apply DBSCAN. Highlight
outliers in the plot
'''
# Generate synthetic data using make_moons
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=0.1, random_state=42)

# Apply DBSCAN
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Identify outliers (points with label -1)
outliers = X[labels == -1]
core_points = X[labels != -1]

# Plot the results
plt.figure(figsize=(8, 6))
# Plot core points and clusters
plt.scatter(core_points[:, 0], core_points[:, 1], c=labels[labels != -1],
            cmap='viridis', label='Clustered Points')
# Highlight outliers in red
plt.scatter(outliers[:, 0], outliers[:, 1], c='red', marker='x', s=100,
            label='Outliers')
plt.title('DBSCAN Clustering on Make_Moons Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
'''
Q29. Create synthetic data using make_blobs and evaluate silhouette scores for
k = 2 to 5. Display as a bar chart
'''
#generate synthetic datasets
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=1000,n_features=2,centers=3,random_state=1)

#kmeans
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2)
kmeans.fit(x)
labels=kmeans.labels_

#silhouette score
from sklearn.metrics import silhouette_score
wcss=[]
for k in range(2,6):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

#bar plot
plt.figure(figsize=(8, 6))
plt.bar(range(2, 6), wcss)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
'''
Q30. Load the Iris dataset and use hierarchical clustering to group data. Plot a
dendrogram with average linkage
'''
#datasets
from sklearn.datasets import load_iris
data=load_iris()
x=data.data
y=data.target

#scalling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

#clustering
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=3,linkage='average')
cluster.fit(x_scaled)

#pca since here 4 feature
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_scalled=pca.fit_transform(x_scaled)

#dendrogram
import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20,7))
plt.title('dendogram')

sc.dendrogram(sc.linkage(pca_scalled,method='average'))

plt.xlabel('sample data')
plt.ylabel('euclidean distance')
plt.show()
```

```python
'''
Q31. Generate synthetic data with overlapping clusters using make_blobs, then
apply K-Means and visualize with decision boundaries
'''
#generate datasets
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=1000,n_features=2,centers=3,cluster_std=2.5,
               random_state=1)

#scalling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

#applying kmeans
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(x_scaled)

#decision boundary
from sklearn.inspection import DecisionBoundaryDisplay
disp=DecisionBoundaryDisplay.from_estimator(
    estimator=kmeans,
    X=x_scaled,
    response_method='predict',
    plot_method='pcolormesh',
    cmap='viridis',
    alpha=0.3
)

disp.ax_.scatter(x_scaled[:,0],x_scaled[:,1],c=kmeans.labels_,cmap='viridis',s=50)
disp.ax_.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                 c='red', s=200, marker='X', label='Centroids')
disp.ax_.set_title("K-Means Clustering with Decision Boundaries")
disp.ax_.legend()
plt.show()
```

```python
'''
Q44. Load the Iris dataset and plot the Silhouette Coefficient for each sample
after KMeans clustering
'''
#load datasets
from sklearn.datasets import load_iris
data=load_iris()
x=data.data
y=data.target

#scalling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

#clustering with Kmeans
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(x_scaled)
labels=kmeans.labels_

#silhouette score
from sklearn.metrics import silhouette_score,silhouette_samples
sample=silhouette_samples(x_scaled,labels)
score=silhouette_score(x_scaled,labels)

plt.figure(figsize=(10, 6))
y_lower = 10
for i in range(3):
    ith_cluster_silhouette_vals = sample[labels == i]
    ith_cluster_silhouette_vals.sort()

    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.nipy_spectral(float(i) / 3)
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_vals,
                      facecolor=color, edgecolor=color, alpha=0.7)

    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, f'Cluster {i}')
    y_lower = y_upper + 10  # spacing between clusters

plt.axvline(x=score, color="red", linestyle="--", label=f'Avg Silhouette Score: {score:.2f}')
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Clustered Samples")
plt.title("Silhouette Plot for KMeans Clustering (Iris Dataset)")
plt.legend()
plt.grid(True)
plt.show()
```
